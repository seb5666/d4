```python
from blackjack import BlackJack
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#build Q-table
num_actions = 2 #Hit or Stick
num_counts = 21-12+1 #Player count can be 12-21
num_dealer_card = 10#Ace-10 (Jack, Queen and King are really the same as a 10)
num_usable_ace = 2 #Either True or False

Q = np.zeros((num_counts, num_dealer_card, num_usable_ace, num_actions))
print("Size of Q-table: {}".format(Q.size))
```

```python
action_table = ["hit", "stick"]

def map_state(state):
    (player_count, dealer_card, usable_ace) = state
    #treat jack, queen and king as 10
    if dealer_card > 9:
        dealer_card = 9
    return (player_count - 12, dealer_card, int(usable_ace))
    
def greedy(state):
    return np.argmax(Q[state])

def ϵ_greedy(state, ϵ = 0.1):
    return np.random.choice(len(action_table)) if (np.random.rand() < ϵ) else greedy(state)
```

```python
def play_optimally():
    game = BlackJack()
    while(not(game.game_ended)):
        action = greedy(map_state(game.get_state()))
        game.take_action(action_table[action])
    return game.game_outcome
```

Using my implementation of blackjack

```python
iters = 5000000
α = 0.1
λ = 1
    
results = []

ϵ_0 = 0.1
ϵ_end = 0.001
ϵ = ϵ_0

α_0 = 0.1
α_end = 0.0001
α = α_0

for i in range(iters):
    game = BlackJack()
    current_state = map_state(game.get_state())
    while(not(game.game_ended)):
        action = ϵ_greedy(current_state, ϵ = ϵ)
        (next_state, reward) = game.take_action(action_table[action])
        next_state = map_state(next_state)

    
        #update Q-table     
        new_estimate = reward
        if not(game.game_ended):
            new_estimate += λ * Q[next_state][greedy(next_state)] 
        Q[current_state][action] = (1 - α) * Q[current_state][action] +  α * new_estimate
        
        current_state = next_state
        
    if i % 10000 == 0:
        s = 0
        for j in range(1000):
            s += play_optimally()
#         ϵ = ϵ_0 - i * (ϵ_0 - ϵ_end)/iters
        α = α_0 - i * (α_0 - α_end)/iters
        results.append(s/1000)
        print("Iteration {}, α={}, ϵ={},\t {}".format(i, np.round(α, 4), np.round(ϵ, 4), results[-1]))
plt.figure()
plt.plot(results)
```

```python
dealer_cards = np.arange(10)
def plot_policy(usable_ace, title = ""):
    values = np.argmax(Q[:,:,int(usable_ace)], axis=-1).transpose()
    print(values)
    min_sticks = np.array([np.argwhere(value > 0)[0] for value in values]).reshape(-1)
    plt.step(dealer_cards+1, min_sticks+11.5, where="mid")
    plt.xlabel("Dealer card")
    plt.ylabel("Player sum")
    plt.xlim(1, 10)
    plt.ylim(12,21)
    plt.title(title)
plot_policy(0, title="No usable ace")
plt.figure()
plot_policy(1, title="Usable ace")
```
